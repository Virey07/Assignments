Scipy: We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below. Task: 1.fitting it to the periodic function 2.plot the fit Data 
##### Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25 
##### Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
temp_max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
temp_min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18
```

```python
def yearly_temps(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))
    
months = np.arange(12)
res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      temp_max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      temp_min, [-40, 20, 0])



days = np.linspace(0, 12, num=365)
plt.figure()
plt.plot(months, temp_max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'y-')
plt.plot(months, temp_min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'g-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```
